```python
pd.options.display.max_rows = 999
a = str((307511-apptrain.isnull().sum())/307511 *100).split("\n")
pd.options.display.max_rows = 16
for i in a:
    if "100.0" not in i:
        print(i) #Printing columns with null values with % of missing data
```

```python
apptrain.describe()
```

```python
# Categorical data needs to be encoded
# I am using label encoding instead of one hot encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
apptrain.dtypes.value_counts() # Objects are categorical data
```

```python
apptrain.select_dtypes("object").columns # all categorical columns
```

```python
apptrain.select_dtypes("object").apply(pd.Series.nunique,axis = 0) # How many unique values are there for each column
```

```python
for i in apptrain.select_dtypes("object").columns:
    try:
        apptrain[str(i)] = le.fit_transform(apptrain[str(i)])
        print(i, "Done")
    except:
        print(i, "Cannot fill due to nan values")
# Deal out with categorical data
```

```python
# Getting columns to fill nan
p = []
for i in a:
    p.append(i.split()[0])
p.pop()
for i in p:
    if apptrain[i].dtype == "object":
        continue
    apptrain[i].fillna(apptrain[i].mean())
    print("Filled NA in col : ",i)
```
